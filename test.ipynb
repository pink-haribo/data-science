```python
# Lec-10
def q1(ser):
  return (ser**2).sum().sum()**0.5

ser = pd.DataFrame([3, 5, 9, 20])
print(q1(ser))

def q2(ser):
  avg = ser.mean()
  std = ser.std()
  std_ser = (ser - avg) / std
  return std_ser

std_ser = q2(pd.Series([-4 , 5, 7, 9 ]))
print(round(std_ser[2], 2))


def q3(ser):
  s_min = ser.min()
  s_max = ser.max()

  return (ser - s_min) / (s_max - s_min)

minmax_ser = q3(pd.Series([-4, 5, 7, 9]))
print(round(minmax_ser[2], 2))
```
